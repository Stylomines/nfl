```python
df_time = df.set_index("date")
```

```python
teams = df_time["team"].unique()
```

```python
mask = df_time["team"]=="Buccaneers"
df_time.loc[mask,["team","pass_completion"]]
```

```python
df_time.select_dtypes(include=['float64']).columns
```

```python
df_time2 = pd.DataFrame(columns = [col for col in df_time.columns])
```

```python
df_time2 = pd.DataFrame(columns = [col for col in df_time.columns])
for team in teams:
    mask = df_time["team"]==team
    df_team_time = df_time[df_time["team"]==team]

    y = df_team_time.loc[:,["pass_completion"]]
    MA7 = y.rolling(window=5)

    df_team_time.assign(pass_completion_MA_5 = MA7.mean())


    df_time2 = pd.concat([df_time2, df_team_time])

df_time2.head()
```

```python
df_time2["pass_completion"][:5]
```

```python
for team in teams:
    mask = df_time["team"]==team
    y = df_time.loc[mask,["team","pass_completion"]]
    MA7 = y.rolling(window=5)
    df_time["pass_completion_MA_5"] = MA7.mean()
```

```python
df_time2["pass_completion_MA_5"].tail(40)
```

```python
pio.templates["jedha"] = go.layout.Template(
    layout_colorway=["#4B9AC7", "#4BE8E0", "#9DD4F3", "#97FBF6", "#2A7FAF", "#23B1AB", "#0E3449", "#015955"]
)
```

```python
y = df_time["pass_completion"]
MA7 = y.rolling(window=7)
df_time["pass_completion_MA_7"] = MA7.mean()

fig = px.line(x=df_time.index, y=df_time["pass_completion"])
fig.add_trace(px.line(x=df_time.index, y=df_time["pass_completion_MA_7"]).data[0])
fig['data'][1]['line']['color']="#4BE8E0"
fig.show()
```
