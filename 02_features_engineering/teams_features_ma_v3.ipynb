## Import libraries

```python
import pandas as pd
import os
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import numpy as np
```

# import file

```python
df = pd.read_csv("../04_datasets/OLD/nfl_team_rating_date.csv")
```

# Calculate Moving Average by average

```python
df_time = df.set_index("date")
df_time.sort_values(by=["team","date"], inplace = True)

df_team_ma = pd.DataFrame(columns = [col for col in df_time.columns])
col_ma = df_time.select_dtypes(include=['float64']).columns
    
teams = df_time["team"].unique()

for team in teams:

    mask = df_time["team"]==team
    df_team_time = df_time[df_time["team"]==team]
    
    for col in col_ma:
    
        y = df_team_time.loc[:,[col]]
        MA5 = y.rolling(window=5)
    
        name_col = col + "_MA_"+ "5"
    
        df_team_time[name_col] = MA5.mean()
  
    df_team_ma = pd.concat([df_team_ma, df_team_time])
    
df_team_ma.reset_index(inplace=True)
```

Save the last Moving Average by team

```python
df_team_topredict = pd.DataFrame(columns = [col for col in df_team_ma.columns])

for team in teams:
    
    mask = df_team_ma["team"]==team
    df_team_time = df_team_ma[df_team_ma["team"]==team]
    df_team_last_time = pd.DataFrame(df_team_time.iloc[-1,:]).T 
    df_team_topredict = pd.concat([df_team_topredict, df_team_last_time])

df_team_topredict.to_csv('../04_datasets/nfl_team_topredict.csv')  
```

Move the moving average of one row

```python
list_col_ma = [col for col in df_team_ma.columns if "_MA_" in col]

for col in list_col_ma :
    
    for i in range(len(df_team_ma)-1,1,-1):
        if df_team_ma["team"][i] == df_team_ma["team"][i-1]:
            df_team_ma[col][i]= df_team_ma[col][i-1]
    
df_team_ma.to_csv('../04_datasets/nfl_team_ma_v2.csv')
```
