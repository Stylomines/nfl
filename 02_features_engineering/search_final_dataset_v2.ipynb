```python
import pandas as pd
```

#### MERGE MASTER FILE (SCORES_PREP) WITH THE MOVING AVERAGE & TIME DELTA FEATURES

##### Step 1: retrieving and cleaning the files

```python
# reading csv files
data_master = pd.read_csv('../04_datasets/nfl_dataset_2.csv') # the master file
data1 = pd.read_csv('../04_datasets/nfl_team_ma.csv') # the file with the moving average + the time delta
```

```python
data_master = data_master.drop(columns=['Unnamed: 0'], axis =0)
data_master.head()
```

```python
data1 = data1.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
data1 = data1.rename(columns = {'index':'game_date'})
data1
```

```python
len(data1)
```

```python
print(data1.columns.tolist())
```

drop columns no Moving Average

```python
data1 = data1.drop(columns=['pass_completion', 'pass_yds', 'pass_avg', 'pass_td', 'pass_int', 'sacks', 'qbr', 'rtg', 'rush_car',
                    'rush_yds', 'rush_avg', 'rush_td', 'rush_long', 'receptions', 'rec_yds', 'rec_avg', 'rec_td', 'rec_long',
                    'rec_tgs', 'fumbles', 'fumbles_lost', 'fumbles_rec', 'defense_tot', 'defense_solo', 'defense_sacks',
                    'defense_tfl', 'defense_pd', 'defense_qb_hits', 'defense_td', 'interceptions', 'interceptions_yds',
                    'interceptions_td', 'kicks_return_no', 'kicks_return_yds', 'kicks_return_avg', 'kicks_return_long',
                    'kicks_return_td', 'punt_return_no', 'punt_return_yds', 'punt_return_avg', 'punt_return_long',
                    'punt_return_td', 'kicking_pct', 'kicking_long', 'kicking_pts', 'punting_no', 'punting_yds',
                    'punting_avg', 'punting_tb', 'punting_in_20', 'punting_long'], axis = 1)
```

import win_streak

```python
data2 = pd.read_csv('../04_datasets/teams_win_streak.csv')
```

```python
data2.head()
```

```python
data2 = data2.drop(columns=['Unnamed: 0','season','week'], axis=1)
```

```python
len(data2)
```

Merge moving average and win_streak

```python
data3 = pd.merge(data2, data1,
                left_on=['idgame','team'],
                right_on = ['idgame','team'],
                how='left')
data3.head(10)
```

```python
len(data3)
```

##### Step 2: merging the files

merge on idgame & awayteam

```python
# using merge function by setting how='left' to keep in line with the master file
output1 = pd.merge(data_master, data3, 
                   left_on=['idgame','awayteam'],
                   right_on = ['idgame','team'],
                   how='left')

# displaying result
output1.head()
```

```python
len(output1)
```

```python
len(output1.columns)
```

merge on idgame & hometeam

```python
output2 = pd.merge(data_master, data3, 
                   left_on=['idgame','hometeam'],
                   right_on = ['idgame','team'],
                   how='left')

# displaying result
output2.head()
```

```python
len(output2)
```

```python
print(output2.columns.tolist())
```

merge away team & home team

```python
output3 = pd.merge(output1, output2, 
                   left_on= 'idgame',
                   right_on = 'idgame',
                   suffixes= ("_away","_home"),
                   how='inner')

# displaying result
output3.head()
```

```python
len(output3)
```

```python
print(output3.columns.tolist())
```

```python
len(output3.columns.tolist())
```

creat new columns delta_day_away & delta_day_home

```python
output3["delta_day_away"] = output3["delta_match_away"].apply(lambda x : str(x).split("days")[0])
output3["delta_day_home"] = output3["delta_match_home"].apply(lambda x : str(x).split("days")[0])
```

#### MERGE MASTER DATAFRAME (output3) WITH THE GAME INFO FEATURES

```python
# the file with the different info on the games (stadium, weather, etc.)
data4 = pd.read_csv('../04_datasets/OLD/games_info_features.csv') 
```

```python
data4.head()
```

```python
 # using merge function by setting how='left' to keep in line with the master file
output4 = pd.merge(output3, data4, 
                   on='idgame', 
                   how='left')
  
# displaying result
output4.head()
```
